Данные необходимо подготовить для обучения модели. Это включает в себя scaling признаков и разделение на обучающую и тестовую выборки.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
feature_columns = [col for col in data.columns if col.startswith('feature_')]
data[feature_columns] = scaler.fit_transform(data[feature_columns])
```

```python
data.head()
```

```python
query_ids = data['query_id'].unique()
train_ids, test_ids = train_test_split(query_ids, test_size=0.2, random_state=42)
```

```python
train_data = data[data['query_id'].isin(train_ids)]
test_data = data[data['query_id'].isin(test_ids)]

print("Train data shape:", train_data.shape)
print("Test data shape:", test_data.shape)
```

# Part 2: Training LambdaMART

```python
import lightgbm as lgb
from sklearn.metrics import ndcg_score
import numpy as np

# Prepare the datasets for LightGBM
train_group = train_data.groupby('query_id').size().to_numpy()
test_group = test_data.groupby('query_id').size().to_numpy()

# Features and labels
X_train = train_data[feature_columns]
y_train = train_data['rank']
X_test = test_data[feature_columns]
y_test = test_data['rank']
```

```python
ranker = lgb.LGBMRanker(
    objective='lambdarank',
    metric='ndcg',
    learning_rate=0.1,
    n_estimators=100,  
    num_leaves=31,    
    seed=42,
    verbose=10,
    early_stopping_rounds=50
)
```

```python
ranker.fit(
    X_train, y_train,
    group=train_group,
    eval_set=[(X_test, y_test)],
    eval_group=[test_group],
    eval_at=[5, 10, 20]
)
```

```python
test_pred = ranker.predict(X_test)
```

```python
def calculate_ndcg(y_true, y_score, k=None):
    true_relevance = np.asarray([y_true])
    scores = np.asarray([y_score])
    if k is None:
        ndcg = ndcg_score(true_relevance, scores)
    else:
        ndcg = ndcg_score(true_relevance, scores, k=k)
    return ndcg
```

```python
ndcg_at_5 = calculate_ndcg(y_test, test_pred, k=5)
print("NDCG@5:", ndcg_at_5)
```

Наша LambdaMART модель с определенными параметрами выбила NDCG@5 = 0.958. 

# Part 3: Hyperparameter tuning for LambdaMART

Ищем лучшие гиперпараметры для модели, оптимизируя NDCG@5.

Стоит отметить, что так как мы используем metric='ndcg', то мы не можем воспользоваться GridSearchCV прямо из sklearn.model_selection.

Поэтому мы напишем свой кастомный подбор гиперпараметров.

```python
learning_rates = [0.01, 0.05, 0.1]
n_estimators_list = [50, 100, 200]
num_leaves_list = [15, 31, 50]
min_child_samples_list = [10, 20, 30]

best_ndcg_score = 0
best_params = {}

for learning_rate in learning_rates:
    for n_estimators in n_estimators_list:
        for num_leaves in num_leaves_list:
            for min_child_samples in min_child_samples_list:
                ranker = lgb.LGBMRanker(objective='lambdarank', metric='ndcg', 
                                         learning_rate=learning_rate, 
                                         n_estimators=n_estimators, 
                                         num_leaves=num_leaves,
                                         min_child_samples=min_child_samples,
                                         seed=42,
                                         verbose=10,
                                         early_stopping_rounds=50)
                ranker.fit(X_train, y_train, group=train_group, eval_set=[(X_test, y_test)], 
                           eval_group=[test_group], eval_at=[5, 10, 20])
                
                test_pred = ranker.predict(X_test)
                ndcg_at_5 = calculate_ndcg(y_test, test_pred, k=5)
                ndcg_at_10 = calculate_ndcg(y_test, test_pred, k=10)
                ndcg_at_20 = calculate_ndcg(y_test, test_pred, k=20)
                
                if ndcg_at_5 > best_ndcg_score:
                    best_ndcg_score = ndcg_at_5
                    best_params = {'learning_rate': learning_rate,
                                   'n_estimators': n_estimators,
                                   'num_leaves': num_leaves,
                                   'min_child_samples': min_child_samples}
                
print("Best Parameters:", best_params)
print("Best NDCG@5 Score:", best_ndcg_score)
```

Как показал наш процесс подбора гиперпараметров, NDCG@5 максимальный при данном наборе гиперпараметров: 

Best Parameters: {'learning_rate': 0.05, 'n_estimators': 100, 'num_leaves': 15, 'min_child_samples': 10}

Best NDCG@5 Score: 0.9671987306219144

Теперь натренируем нашу модель с данными гиперпараметрами.

```python
# Train the model with the best parameters
best_ranker = lgb.LGBMRanker(objective='lambdarank', metric='ndcg', 
                              learning_rate=0.05, n_estimators=100, 
                              num_leaves=15, min_child_samples=10,
                              seed=42, verbose=10, early_stopping_rounds=50)

best_ranker.fit(X_train, y_train, group=train_group, eval_set=[(X_test, y_test)], 
                eval_group=[test_group], eval_at=[5, 10, 20])

test_pred = best_ranker.predict(X_test)
```

# Part 4: Final model evaluation

```python
# Calculate the metrics
ndcg_at_5 = calculate_ndcg(y_test, test_pred, k=5)
ndcg_at_10 = calculate_ndcg(y_test, test_pred, k=10)
ndcg_at_20 = calculate_ndcg(y_test, test_pred, k=20)
ndcg = calculate_ndcg(y_test, test_pred)

print("NDCG@5:", ndcg_at_5)
print("NDCG@10:", ndcg_at_10)
print("NDCG@20:", ndcg_at_20)
print("NDCG:", ndcg)
```

Модель показала хорошие результаты по всем метрикам.
